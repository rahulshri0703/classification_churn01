```python
class EDA():
    
    def __init__(self,df):
        self.df = df
        
    def plot(self,col,kind='kdeplot',hue=None,y=False,scale='linear',figsize=(9,4)):
        
        
        n = len(col)

        plt.figure(figsize=figsize)
        for i in range(1,n):

            plt.subplot(1,n,i)
            m = getattr(sns,kind)

            if y == True:
                
                if (kind == 'boxplot' or kind == 'violinplot'):
                    m(data=self.df,y=col[i],x=hue)
                
                else:
                    m(data=self.df,y=col[i],hue=hue)

                if scale == 'log':
                    plt.yscale('log')
            else:
                
                if (kind == 'boxplot' or kind == 'violinplot'):
                    
                    m(data=self.df,x=col[i],y=hue)
                
                else:
                    m(data=self.df,x=col[i],hue=hue)

                if scale == 'log':
                    plt.xscale('log')

        plt.tight_layout()
        
    
    def plot_bivariate(self,col,y,kind='barplot',scale='linear',figsize=(6,7),hue=None):
        
        n = len(col)

        plt.figure(figsize=figsize)
        for i in range(1,n):

            plt.subplot(n,1,i)
            m = getattr(sns,kind)
            
            if kind == 'regplot':
                m(data=self.df,y=y,x=col[i])
              
            else:
                m(data=self.df,y=y,x=col[i],hue=hue)

            if scale == 'log':
                plt.yscale('log')
                
        plt.tight_layout()
        
    def describe(self,col,perc=[0.15,0.25,0.5,0.75,0.9,0.95]):
        
        d = pd.DataFrame()
        d['dtypes'] = self.df[col].dtypes
        d['null'] = self.df[col].isnull().mean()
        d['nunique'] = self.df[col].nunique()
        d['count'] = self.df[col].count()
        
        print('\n',self.df[col].describe(percentiles=perc))
        print('\n',d)
        
    def correlation(self,col,tol=0.7):
        
        corr = self.df[col].corr()
        c = corr.unstack().reset_index()
        c['corr']=abs(c[0])
        c= c.sort_values(by='corr',ascending=False)
        c = c[(c[0]!=1)&(c['corr']>tol)]
        c.style.backgound_gradient(cmap=Reds)
        print(c)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

def try_diff_model(model,X_train,y_train,X_test,y_test,columns,confusion=True,
                   summary=False,vif=False):
    
    baseline_model = model # `weight of class` balancing technique used
    baseline_model = baseline_model.fit(X_train[columns], y_train)

    y_train_pred_proba = baseline_model.predict_proba(X_train[columns])[:,1]
    y_test_pred_proba  = baseline_model.predict_proba(X_test[columns])[:,1]

    y_train_pred = baseline_model.predict(X_train[columns])
    y_test_pred  = baseline_model.predict(X_test[columns])

    y_train_pred_proba = pd.Series(y_train_pred_proba,index = X_train[columns].index, ) # converting test and train to a series to preserve index
    y_test_pred_proba = pd.Series(y_test_pred_proba,index = X_test[columns].index)
    
   
    train_matrix = classification_report(y_train, y_train_pred)
    test_matrix = classification_report(y_test, y_test_pred)
    print('\nClassification report for test: \n', test_matrix)
    
   
    train_matrix = roc_auc_score(y_train, y_train_pred_proba)
    print('ROC for train:\n', train_matrix,'\n')
    test_matrix = roc_auc_score(y_test, y_test_pred_proba)
    print('\nROC for test: \n', test_matrix,'\n')
    
    if confusion == True:
        train_matrix = confusion_matrix(y_train, y_train_pred)
        print('Confusion Matrix for train:\n', train_matrix)
        test_matrix = confusion_matrix(y_test, y_test_pred)
        print('\nConfusion Matrix for test: \n', test_matrix)
    
    if summary == True:
        
        logr = sm.GLM(y_train,
              (sm.add_constant(X_train[columns])),
              family = sm.families.Binomial())
        logr_fit = logr.fit()
        print('\n',logr_fit.summary())
        
    if vif ==True:
            vif = pd.DataFrame()
            vif['Features'] = columns
            vif['VIF'] = [variance_inflation_factor(X_train[columns].values, i) 
                          for i in range(X_train[columns].shape[1])]
            vif['VIF'] = round(vif['VIF'], 2)
            vif = vif.set_index('Features')
            vif['P-value'] = round(logr_fit.pvalues,4)
            vif = vif.sort_values(by = ["VIF",'P-value'], ascending = [False,False])
            print('\n',vif)
        
    
    return y_train_pred,y_test_pred,y_train_pred_proba,y_test_pred_proba
```

```python
#Baseline Performance

from sklearn.linear_model import LogisticRegression

output = try_diff_model(model = LogisticRegression(random_state=100, class_weight='balanced'
                                                  ,max_iter=600),
                       X_train = X_train,y_train = y_train,
                        X_test = X_test,y_test = y_test,
                        columns=X_train.columns)
```
